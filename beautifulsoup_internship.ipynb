```python
#question1
import requests
from bs4 import BeautifulSoup
import pandas as pd

# website url for IMBD page
web_url = "https://www.imdb.com/list/ls056092300/"

#apply get request to fetch html url
response = requests.get(web_url)

# getting the raw html content
soup = BeautifulSoup(response.content, 'html.parser')

# filtering the movie containers(which has the names, ratings and year of release)
movie_containers = soup.find_all('div', class_='lister-item-content')

#create emopty list in order to store the extracted details from the containers
movie_names = []
movie_ratings = []
year_of_release = []

# using for loop iterate the movie containers each time to fetch the required details
for container in movie_containers:
    # name of the movie
    name = container.h3.a.text.strip()
    movie_names.append(name)

    # Movie_rating
    rating = container.find('span', class_='ipl-rating-star__rating').text.strip()
    movie_ratings.append(rating)

    # year of the release
    year = container.find('span', class_='lister-item-year').text.strip('()')
    year_of_release.append(year)

# creating the data frame to store the extracted details
df = pd.DataFrame({
    'Name': movie_names,
    'Rating': movie_ratings,
    'Year of Release': year_of_release
})

# print the DataFrame
print(df)
```

```python
#question2
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import time

web_url = "https://peachmode.com/search?q=bags"

# Using Chrome webdriver. Make sure to install the appropriate webdriver for your browser.
driver = webdriver.Chrome()

driver.get(web_url)

# Sleep for a few seconds to ensure the page is fully loaded
time.sleep(5)

soup = BeautifulSoup(driver.page_source, 'html.parser')

product_containers = soup.find_all('div', class_="product-inner")

prod_names = []
prod_price = []
prod_discounts = []

for container in product_containers:
    name = container.find('div', class_="product-title").text.strip()
    prod_names.append(name)
    
    price = container.find('span', class_="discounted_price").text.strip()
    prod_price.append(price)
    
    
    discount = container.find('span', class_="price st-money money done").text.strip()
    prod_discounts.append(discount)
    
driver.quit()

df = pd.DataFrame({
    'Product Name': prod_names,
    'Price': prod_price,
    'Discount': prod_discounts
})

print(df)
```

```python
#question 4
import requests
from bs4 import BeautifulSoup

# Send a GET request to the Patreon page
url = "https://www.patreon.com/coreyms"
response = requests.get(url)

# Parse the HTML content
soup = BeautifulSoup(response.content, "html.parser")

# Find all the posts
posts = soup.find_all("div", attrs={"data-tag": "post-card"})

for post in posts:
    # Extract heading
    heading = post.find("span", attrs={"data-tag": "post-title"}).text.strip()
    
    # Extract date
    date = post.find("a", attrs={"data-tag": "post-published-at"}).text.strip()
    
    # Extract content
    content = post.find("div",class_="hgUaki").text.strip()
    
    # Extract YouTube video link
    youtube_link = post.find("svg", attrs={"data-tag": "IconBrandYoutube"})
    if youtube_link:
        youtube_url = youtube_link["href"]
        # Assuming the YouTube link is in the format: https://www.youtube.com/watch?v=<video_id>
        video_id = youtube_url.split("posts/")[1]
        youtube_video_link = f"/posts/{video_id}"
        #href="/posts/python-youtube-37666379"
    else:
        youtube_video_link = "No YouTube video link found"
        
    
    
    # Extract likes for the YouTube video (if available)
    likes = post.find("span", attrs={"data-tag": "like-count"})
    if likes:
        likes_count = likes.text.strip()
    else:
        likes_count = "No likes count available"
    
    # Print the details
    print("Heading:", heading)
    print("Date:", date)
    print("Content:", content)
    print("YouTube Video Link:", youtube_video_link)
    print("Likes Count:", likes_count)
    
```

```python
#question6
import requests
from bs4 import BeautifulSoup

# Define the URL
url = "https://www.cnbc.com/world/?region=world"

# Send a GET request to the URL
response = requests.get(url)

# Parse the HTML content
soup = BeautifulSoup(response.content, "html.parser")

# Find all the news articles
articles = soup.find_all("div", class_="LatestNews-container")

# Iterate over each article and extract details
for article in articles:
    # Extract heading
    #heading = article.find("h3").text.strip()
    heading = article.find("a", class_="LatestNews-headline").text.strip()
    
    # Extract date
    date = article.find("time").text.strip()
    
    # Extract news link
    link = "https://www.cnbc.com" + article.find("a")["href"]
    
    # Print the details
    print("Heading:", heading)
    print("Date:", date)
    print("Link:", link)
    print("-" * 50)
```

```python
#question7
import requests
from bs4 import BeautifulSoup

# Define the URL
url = "https://www.keaipublishing.com/en/journals/artificial-intelligence-in-agriculture/most-downloaded-articles/"

# Send a GET request to the URL
response = requests.get(url)

# Parse the HTML content
soup = BeautifulSoup(response.content, "html.parser")

# Find all the articles
articles = soup.find_all("div", class_="article-listing")

# Iterate over each article and extract details
for article in articles:
    # Extract paper title
    title = article.find("h2", class_="h5 article-title").text.strip()
    
    # Extract date
    date = article.find("p", class_="article-date").text.strip()
    
    # Extract author
    author = article.find("p", class_="article-authors").text.strip()
    
    # Print the details
    print("Paper Title:", title)
    print("Date:", date)
    print("Author:", author)
    print("." * 50)
```
